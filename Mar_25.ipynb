## Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python

%matplotlib inline
# Filter out non-numeric values in the Price column
filtered_df = df[df['Price'] != 'Everyone']
filtered_df['Price'] = filtered_df['Price'].str.replace('$', '').astype(float)

# Group the data by app type and calculate the average prices
avg_prices = filtered_df.groupby('Type')['Price'].mean().reset_index()

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(avg_prices['Type'], avg_prices['Price'])
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average App Prices by Type')
plt.show()
```

## Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?


```python
import pandas as pd

# Read the Google Play Store dataset into a DataFrame
# df = pd.read_csv("googleplaystore.csv")

# Get the list of app categories
categories = df["Category"].tolist()

# Create a dictionary to store the number of apps in each category
num_apps = {}
for category in categories:
    num_apps[category] = len(df[df["Category"] == category])

# Create a dictionary to store the number of installs in each category
num_installs = {}
for category in categories:
    num_installs[category] = df[df["Category"] == category]["Installs"].sum()

# Create a dictionary to store the average rating in each category
avg_rating = {}
for category in categories:
    avg_rating[category] = df[df["Category"] == category]["Rating"].mean()

# Print the most popular app categories
for category, num_apps in num_apps.items():
    print(f"{category}: {num_apps}")

# Print the categories with the highest number of installs
for category, num_installs in num_installs.items():
    if num_installs == max(num_installs.values()):
        print(f"{category}: {num_installs}")

# Print the categories with the highest average rating
for category, avg_rating in avg_rating.items():
    if avg_rating == max(avg_rating.values()):
        print(f"{category}: {avg_rating}")
```
